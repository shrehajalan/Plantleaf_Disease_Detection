```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```python
import pathlib
data_dir="./train"
data_dir = pathlib.Path(data_dir)
data_dir
```

```python
corn_images_dict = {
    'healthy':list(data_dir.glob('Healthy/*')),
    'blight':list(data_dir.glob('Blight/*')),
    'Common_Rust':list(data_dir.glob('Common_Rust/*')),
}
corn_labels_dict = {
    'healthy': 0,
    'blight': 1,
    'Common_Rust': 2,
}
```

```python
hel=len(corn_images_dict['healthy'])
bli=len(corn_images_dict['blight'])
cr=len(corn_images_dict['Common_Rust'])

print(f"number of photo of healthy  ={hel}   number of photo of early blight ={bli}   number of photo of lateblind ={cr} ")
```

```python
import random
fig=plt.figure(1,figsize=(15,9))
fig.suptitle('photo', fontsize=25)
n=0
for i in range(15):
    n+=1
    if n<=5:
        random_im=random.choice(corn_images_dict['healthy'])
        k=plt.imread(random_im)
        plt.subplot(4, 4, n)
        p=plt.imshow(k)
        l=np.shape(k)
        plt.title(f"Healthy{l}")
        plt.axis("off")
    elif n<=10:
        random_im=random.choice(corn_images_dict['blight'])
        k=plt.imread(random_im)
        plt.subplot(4, 4, n)
        p=plt.imshow(k)
        l=np.shape(k)
        plt.title(f"blight{l}")
        plt.axis("off")
    elif n<=15:
        random_im=random.choice(corn_images_dict['Common_Rust'])
        k=plt.imread(random_im)
        plt.subplot(4, 4, n)
        p=plt.imshow(k)
        l=np.shape(k)
        plt.title(f"Common_Rust{l}")
        plt.axis("off")
```

```python
import cv2
from tqdm.auto import tqdm
X, y = [], []
count=0

for disease_name, images in tqdm(corn_images_dict.items()):
    for image in images:
        img = cv2.imread(str(image))
        resized_img = cv2.resize(img,(256,256))
        if resized_img.shape==(256,256,3):
            X.append(resized_img)
            y.append(corn_labels_dict[disease_name])
            count+=1
        
```

```python
print(f'Total image is: {count}')
print(f"shape of y is :{np.shape(y)}")
print(f"shape of X is :{np.shape(X)}")
```

```python
x=np.array(X)
y=np.array(y)
```

```python
x=x/255
```

```python
item={
     0:'Healthy',
     1:'Blight',
     2:'Common_Rust',
}


plt.figure(1,figsize=(15,9))
plt.suptitle('photo after append in x',fontsize=25)
n=0
for i in range(16):
    n+=1
    a=random.randint(0,3613)
    plt.subplot(4, 4, n)
    p=plt.imshow(x[a])
    l=np.shape(x[a])
    ko=y[a]
    plt.title(f"{item[ko]} ,{l}")
    plt.axis("off")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y,random_state=0)
x_train, x_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.15, shuffle=True)
```

```python
print(f'x_train shape={np.shape(x_train)}')
print(f'y_train shape = {np.shape(y_train)}')
```

```python
print(f'x_train shape={np.shape(x_train)}')
print(f'x_test shape={np.shape(X_test)}')
print(f'x_val shape={np.shape(x_val)}')
```

```python
num_classes = 3

model = Sequential([
  layers.Conv2D(16, 3, padding='same', activation='relu',input_shape=(256,256,3)),
  layers.MaxPooling2D((2,2)),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D((2,2)),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D((2,2)),
  layers.Conv2D(80, 3, padding='same', activation='relu'),
  layers.MaxPooling2D((2,2)),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(256, activation='relu'),
  layers.Dense(num_classes,activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
history=model.fit(x_train, y_train, validation_data=(x_val,y_val), epochs=10)
```

```python
loss_train=history.history['loss']
acc_train=history.history['accuracy']
loss_val=history.history['val_loss']
acc_val=history.history['val_accuracy']
epochs=[i for i in range(1,11)]
```

```python
plt.plot(epochs, loss_train, 'g',label='Training loss')
plt.plot(epochs, loss_val, 'b', label='validation loss')
```

```python
plt.plot(epochs, acc_train, 'g',label='Training accuracy')
plt.plot(epochs, acc_val, 'b', label='validation accuracy')
```

```python
predictions = model.predict(X_test)
predictions
```

```python
score=[np.argmax(ele) for ele in predictions]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,score))
```

```python
test_confusion_matrix = tf.math.confusion_matrix(y_test, score)
ax = sns.heatmap(test_confusion_matrix, annot = True, fmt ='g')
ax.set(xlabel = 'Actual values', ylabel = 'Predicted values')
plt.show()
```
