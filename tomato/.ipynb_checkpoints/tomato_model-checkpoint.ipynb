```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```python
import pathlib
data_dir="./train"
data_dir = pathlib.Path(data_dir)
data_dir
```

```python
diseased_images_dict = {
    'Bacterial':list(data_dir.glob('Tomato___Bacterial_spot/*')),
    'Early_blight':list(data_dir.glob('Tomato___Early_blight/*')),
    'Healthy':list(data_dir.glob('Tomato___healthy/*')),
    'Late_blight':list(data_dir.glob('Tomato___Late_blight/*')),
    'Leaf_Mold':list(data_dir.glob('Tomato___Leaf_Mold/*')),
    'Septoria_leaf_spot':list(data_dir.glob('Tomato___Septoria_leaf_spot/*')),
    'Target_Spot':list(data_dir.glob('Tomato___Target_Spot/*')),
    'Yellow_Leaf_Curl_Virus':list(data_dir.glob('Tomato___Tomato_Yellow_Leaf_Curl_Virus/*')),
    'Mosaic_virus':list(data_dir.glob('Tomato___Tomato_mosaic_virus/*')),
    'Two-spotted_spider_mite':list(data_dir.glob('Tomato___Spider_mites Two-spotted_spider_mite/*')),
    
}
diseased_labels_dict = {
   'Bacterial':0,
   'Early_blight':1,
   'Healthy':2,
   'Late_blight':3,
   'Leaf_Mold':4,
   'Septoria_leaf_spot':5,
   'Target_Spot':6,
   'Yellow_Leaf_Curl_Virus':7,
   'Mosaic_virus':8,
   'Two-spotted_spider_mite':9,
   
}
```

```python
for i in diseased_images_dict.keys():
    print(i,"=",len(diseased_images_dict[i]),"images")
```

```python
import cv2
from tqdm.auto import tqdm
X, y = [], []
count=0

for disease_name, images in tqdm(diseased_images_dict.items()):
    for image in images:
        img = cv2.imread(str(image))
        resized_img = cv2.resize(img,(128,128))
        if resized_img.shape==(128,128,3):
            X.append(resized_img)
            y.append(diseased_labels_dict[disease_name])
            count+=1
```

```python
print(f'Total image is: {count}')
print(f"shape of y is :{np.shape(y)}")
print(f"shape of X is :{np.shape(X)}")
```

```python
x=np.array(X)
y=np.array(y)
```

```python
x=x/255
```

```python
import random
item={
   0:'Bacterial',
   1:'Early_blight',
   2:'Healthy',
   3:'Late_blight',
   4:'Leaf_Mold',
   5:'Septoria_leaf_spot',
   6:'Target_Spot',
   7:'Yellow_Leaf_Curl_Virus',
   8:'Mosaic_virus',
   9:'Two-spotted_spider_mite',
}


plt.figure(1,figsize=(15,9))
plt.suptitle('photo after append in x',fontsize=25)
n=0
for i in range(16):
    n+=1
    a=random.randint(0,9999)
    plt.subplot(4, 4, n)
    p=plt.imshow(x[a])
    l=np.shape(x[a])
    ko=y[a]
    plt.title(f"{item[ko]}")
    plt.axis("off")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y,random_state=0)
x_train, x_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.15, shuffle=True)
```

```python
print(f'y_train shape = {np.shape(y_train)}')
print(f'x_train shape={np.shape(x_train)}')
print(f'x_test shape={np.shape(X_test)}')
print(f'x_val shape={np.shape(x_val)}')
```

```python
num_classes = 10

model = Sequential([
  layers.Conv2D(16, 3, padding='same', activation='relu',input_shape=(128,128,3)),
  layers.MaxPooling2D((2,2)),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D((2,2)),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D((2,2)),
  layers.Conv2D(80, 3, padding='same', activation='relu'),
  layers.MaxPooling2D((2,2)),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(256, activation='relu'),
  layers.Dense(num_classes,activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
history=model.fit(x_train, y_train, validation_data=(x_val,y_val), epochs=10)  
```

```python
loss_train=history.history['loss']
acc_train=history.history['accuracy']
loss_val=history.history['val_loss']
acc_val=history.history['val_accuracy']
epochs=[i for i in range(1,11)]
```

```python
plt.plot(epochs, loss_train, 'g',label='Training loss')
plt.plot(epochs, loss_val, 'b', label='validation loss')
```

```python
plt.plot(epochs, acc_train, 'g',label='Training accuracy')
plt.plot(epochs, acc_val, 'b', label='validation accuracy')
```

```python
predictions = model.predict(X_test)
predictions
```

```python
score=[np.argmax(ele) for ele in predictions]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,score))
```

```python
test_confusion_matrix = tf.math.confusion_matrix(y_test, score)
ax = sns.heatmap(test_confusion_matrix, annot = True, fmt ='g')
ax.set(xlabel = 'Actual values', ylabel = 'Predicted values')
plt.show()
```

## Transfer Learning

## RestNet 50

```python
model1 = tf.keras.applications.ResNet50(weights='imagenet',include_top=False,input_shape=(128, 128, 3))

for layer in model1.layers:
   layer.trainable=False
```

```python
x=model1.output
x=tf.keras.layers.GlobalAveragePooling2D()(x)
x=tf.keras.layers.Dense(120,activation='relu')(x)
pred=tf.keras.layers.Dense(10,activation='softmax')(x)

model2 = keras.Model(inputs=model1.input, outputs=pred)
```

```python
model2.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model2.summary()
```

```python
History1=model2.fit(x_train, y_train, validation_data=(x_val,y_val), epochs=10)  
```

## MobileNet

```python
model3=tf.keras.applications.mobilenet.MobileNet(
    include_top=False,
    weights='imagenet',
    input_shape=(128, 128, 3)
)
for layer in model3.layers:
  layer.trainable=False
```

```python
x=model3.output
x=tf.keras.layers.GlobalAveragePooling2D()(x)
x=tf.keras.layers.Dense(128,activation='relu')(x)
pred=tf.keras.layers.Dense(10,activation='softmax')(x)

model4= keras.Model(inputs=model3.input, outputs=pred)
```

```python
model4.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model4.summary()
```

```python
History2=model4.fit(x_train, y_train, validation_data=(x_val,y_val), epochs=10)  
```

```python
loss_train=History2.history['loss']
acc_train=History2.history['accuracy']
loss_val=History2.history['val_loss']
acc_val=History2.history['val_accuracy']
epochs=[i for i in range(1,11)]
```

```python
plt.plot(epochs, loss_train, 'g',label='Training loss')
plt.plot(epochs, loss_val, 'b', label='validation loss')
```

```python
plt.plot(epochs, acc_train, 'g',label='Training accuracy')
plt.plot(epochs, acc_val, 'b', label='validation accuracy')
```

```python
predictions1 = model4.predict(X_test)
predictions1
```

```python
score=[np.argmax(ele) for ele in predictions1]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,score))
```

```python
test_confusion_matrix = tf.math.confusion_matrix(y_test, score)
ax = sns.heatmap(test_confusion_matrix, annot = True, fmt ='g')
ax.set(xlabel = 'Actual values', ylabel = 'Predicted values')
plt.show()
```
